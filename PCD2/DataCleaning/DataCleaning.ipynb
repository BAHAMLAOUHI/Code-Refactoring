# we have noticed some mixed types in our Dataframes
# we will check first which columns have a non numeric values and display them 

```python
numeric_columns = df.select_dtypes(include='number').columns
non_numeric_columns = df.select_dtypes(exclude='number').columns
print("These are the non Numiric columns in the data Set")
print("-----------------------------------------------------------------------------------")
print(non_numeric_columns)
```

# now we will see the types of the non numric values of each column ( metrics)

```python
B=non_numeric_columns[6:]
for i in range(0,len(B)):
    unique_types = df[B[i]].apply(type).unique()
    print("-----------------------------------------------------------------------------------")
    print(f"these are the types of '{B[i]}' '{unique_types}'."  )
```

# first python will know that a float is a float if there is a '.' not a ',' 

# so we need to change it first and then convert all values to number ( strvalues)

```python
# df[non_numeric_columns] = df[non_numeric_columns].apply(lambda x: x.str.replace(',', '.'))
for column in non_numeric_columns:
    if df[column].dtype != int and column not in ['ID', 'Name', 'LongName', 'Parent', 'Component', 'Path']:
        df[column] = df[column].apply(lambda x: str(x).replace(',', '.'))
```

# we will see an example of a mixed type feature ( metric)

```python
print("Unique values in 'PrevCC':", df['PrevCC'].unique())
print("-----------------------------------------------------------------------------------")
print("Unique values in 'CC':", df['CC'].unique())
```

# Now we will change all values to numeric type

```python
# Now we will change all values to numeric type
```

```python
for column in non_numeric_columns:
    if column not in ['ID', 'Name', 'LongName', 'Parent', 'Component', 'Path']:
        try:
            df[column] = pd.to_numeric(df[column], errors='coerce')
        except ValueError:
            # If the conversion fails, it will raise a ValueError
            print(f"Conversion failed for column '{column}'.")
```

# we will see the same example after conversion 


```python
print("Unique values in 'PrevCC':", df['PrevCC'].unique())

print("Unique values in 'Name':", df['Name'].unique())
```

# Step 2: Handling Duplicates ( samples )


```python
df.drop_duplicates(inplace=True)

# Print changes only if there are differences
if not df.equals(df):
    print(" there is some Duplicated samples and removed successfully ")
else: print (" there is no changes to the data set ( No duplicated samples)")
```

```python
columns_without_prev = []

# Find columns without 'Prev' prefix
for col in df.columns:
    if not col.startswith('Prev'):
        columns_without_prev.append(col) 

# Iterate over columns without 'Prev'
for col in columns_without_prev:
    prev_column = 'Prev' + col
    if prev_column not in df.columns and col not in ['Refact_Sum']:
        # If corresponding Prev column does not exist, drop the non-Prev column
        df.drop(col, axis=1, inplace=True)

df
        
    
```

```python
# Extract the column you want to move
feature_to_move = df.pop('Refact_Sum')

# Reinsert the column at the end
df['Refact_Sum'] = feature_to_move

df
```

```python
# Apply a lambda function to map the values
df['Refact_Sum'] = df['Refact_Sum'].apply(lambda x: 1 if x > 0 else x)
```
