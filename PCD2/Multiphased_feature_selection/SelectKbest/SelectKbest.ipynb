```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the Excel file into a DataFrame
excel_file_path = 'C:/Users/baha eddine/Desktop/PCD2/Multiphased_feature_selection/Cross-correlation/DS_after_Correlation_final.xlsx'
df = pd.read_excel(excel_file_path)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
X = df.drop(columns=['Refact_Sum'])
y = df['Refact_Sum']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Scale the input data to a range of [0, 1]
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python

# Initialize SelectKBest with chi2 scoring function
selector = SelectKBest(score_func=chi2, k=24)

# Fit the selector to your data
selector.fit(X_scaled, y)

# Get the indices of the selected features
selected_indices = selector.get_support(indices=True)

# Get the names of the selected features
selected_features = X.columns[selected_indices]

selected_features
```

```python
len(selected_features)
```

```python
# on vas just remplacer le metric de DLOC by LCOM5 metric
final_col=['LCOM5']
for col in selected_features :
    if col not in ['DLOC']:
        final_col.append(col)
        
final_col        
```

```python
# LCOM5: Lack of Cohesion in Methods version 5 
# NL: Nesting Level
# NLE: Nesting Level Else-if
# WMC :Weighted Methods per Class
# CBO: Coupling Between Objects
# CBOI: Coupling Between Objects Classes Inverse 
# NII:Number of incoming invocation 
# NOI: Number of Outgoing invocations 
# RFC :Response For a Class
# CLOC :Commentt Lines of Code 
# TCLOC : Total Commentt Lines of Code 
# LLOC: Logical Lines of Code
# TLLOC: Total Logical Lines of Code
# LOC : Line Of Code 
# TLOC : Total Line Of Code 
# NG : Number of getters 
# NLG : Number of Local getters
# NLM : Number of Local Methods
# NLPM: Number of local public Methods
# NOS: number of statements 
# TNLG: Total Number of local Getters 
# TNLM : Total number of Local Methods 
# TNLPM: Total number of local public Methods
# TNOS : Total Number of Statements
```

```python
for col in df.columns:
    if col not in final_col and col not in ['Refact_Sum']:
        df=df.drop(col,axis=1)
df        
```
