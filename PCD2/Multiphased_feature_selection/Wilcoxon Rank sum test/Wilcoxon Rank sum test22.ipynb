```python
import scipy.stats as stats
```

```python
# Extract 'Prev' and 'Value' columns for each metric and perform Wilcoxon test
alpha = 0.05  # significance level
nb_treated_metrics=0
# nbre 
nb_yes=0
nb_no=0
nb_all_metrics=0
allMetrics=[]
MetricswithPrev=[]

significant_columns = []
non_significant_columns = []

for metric in df.columns:
    nb_all_metrics+=1
    if 'Prev' not in metric:
        allMetrics.append(metric)
        continue  # Skip columns without 'Prev'
    
    # Identify the corresponding 'Value' column
    value_column = metric.replace('Prev', '')
    MetricswithPrev.append(value_column)
    
#     # Convert columns to numeric (assuming both 'Prev' and 'Value' columns have numeric data)
#     df[metric] = pd.to_numeric(df[metric], errors='coerce')
#     df[value_column] = pd.to_numeric(df[value_column], errors='coerce')
    
    # Calculate differences
    differences = df[value_column] - df[metric]
    
    # Perform Wilcoxon signed-rank test
    # Exclude zeros from differences
    non_zero_diff = differences[differences != 0]

    # Perform Wilcoxon signed-rank test
    statistic, p_value = stats.wilcoxon(non_zero_diff, zero_method='pratt', alternative='two-sided')  
    if p_value < alpha:
        print( value_column+" :Yes")
        nb_treated_metrics+=1
        nb_yes+=1
        significant_columns.append(value_column)
    else:
        print(value_column +" :No")
        nb_treated_metrics+=1
        nb_no+=1
        non_significant_columns.append(value_column)
        
print("-----------------------------------------------------------------------------------")
print( "Total treated Metrics :", nb_treated_metrics)   
print("-----------------------------------------------------------------------------------")
print("Total Number of columns in the data set ", nb_all_metrics - nb_treated_metrics)
print("-----------------------------------------------------------------------------------")
```

```python
print("There is",nb_yes ,"Significant Columns:")
print(significant_columns)
print("-----------------------------------------------------------------------------------")


print("-----------------------------------------------------------------------------------")
print("There is",nb_no ,"Non Significant Columns:")
print(non_significant_columns)
print("-----------------------------------------------------------------------------------")


non_treated_metrics = [item for item in allMetrics if item not in MetricswithPrev]
print("Not treated Columns :")
print(non_treated_metrics)
print("-----------------------------------------------------------------------------------")
# 
```

```python
# Remove non-significant columns along with their 'Prev' values from the DataFrame
df = df.drop(columns=non_significant_columns + [f"Prev{col}" for col in non_significant_columns])
```
